

### Elevator Pitch for Python


**Number of public repositories on GitHub matching the topics `python`, `r`, and `stata` as of Thu Aug 1, 2024.**

For nearly 20 years, the 750 series has been providing students with Stata code to help develop concepts taught in class. Due to increasing demand, the series introduced R code as well. The department is clearly catching up with trends, but from the opposite side of the spectrum. I speculate that very soon students will be demanding code in Python. 

### Here’s why:

1. **Versatility and Accessibility**:
   - Python is a general-purpose language that excels in various domains such as data science, machine learning, web development, and automation. Its syntax is straightforward and readable, making it accessible for beginners and efficient for experts.

2. **Rich Ecosystem**:
   - Python boasts an extensive library ecosystem. Packages like `pandas`, `numpy`, `scikit-learn`, and `matplotlib` are essential for data analysis and visualization. `TensorFlow` and `PyTorch` are leading libraries for machine learning.

3. **Industry and Academic Adoption**:
   - Python is the most popular programming language in both academia and industry. Leading tech companies like Google, Facebook, and Netflix rely heavily on Python. In academia, Python is increasingly the language of choice for teaching data science and computational biology.

4. **Community and Support**:
   - Python has a vast and active community. The abundance of tutorials, forums, and conferences ensures that students and researchers can find support and resources easily.

5. **Interoperability**:
   - Python integrates seamlessly with other languages and technologies, making it a versatile tool in multi-language projects. It can be used to glue together workflows involving R, Stata, and even C++ or Java.

6. **Future-Proofing**:
   - As data science and machine learning continue to evolve, Python is at the forefront of innovation. Learning Python equips students with skills that are highly relevant and in demand.

**Conclusion**:
The overwhelming number of public repositories and the growing interest in Python highlight its importance and utility. Adopting Python in our curriculum will not only meet the current demand but also prepare our students for the future. Let's position our department at the cutting edge of this trend.

**Pitch Outline:**

*Subject: Adapting to Gen Z Learners: Embracing Dynamic Educational Tools in Epidemiology*

---

**1. Understanding the Gen Z Graduate Student:**

Today’s graduate students in epidemiology predominantly belong to Generation Z (born 1997-2013). This generation has grown up with smartphones, social media, and instantaneous access to information. They’ve never known a world without dynamic content and are accustomed to highly personalized, interactive experiences. Static educational materials—textbooks, PDFs, and even traditional PowerPoints—do not engage them in the same way. In contrast, interactive web apps fit their learning style and can significantly enhance their understanding of complex epidemiological concepts.

---

**2. Why Web Apps Are the Future of Epidemiology Education:**

- **Active vs. Passive Learning**: Gen Z students learn best through engagement and real-time interaction. A static lecture or reading assignment only scratches the surface. With interactive web apps, students can simulate real-world epidemiological scenarios, test hypotheses, and instantly see the consequences of different variables. For example, simulating outbreaks in different environments helps them visualize the spread of diseases in real-time.
  
- **Customization**: Web apps can be tailored to different learning levels, from beginner to advanced. Graduate students can have the flexibility to dive deeper into more complex areas or repeat basic concepts as needed.

- **Instant Feedback**: Unlike static material, dynamic platforms provide instant feedback, which is key to learning. By engaging with simulations, quizzes, and interactive tools, students can correct misunderstandings immediately, leading to better retention and application of knowledge.

---

**3. Dynamic Tools for the Entire Epidemiology Ecosystem:**

Beyond students, web apps are essential for professionals across the spectrum in public health and epidemiology:

- **For Professors**: Web apps enable professors to monitor progress in real time and offer personalized feedback. Instead of grading static assignments, professors can assess students based on real-time engagement with case studies and data sets in interactive platforms.

- **For Patients and Public Health Officials**: Interactive epidemiological tools can be extended to patients and public health professionals to visualize local outbreaks, predict trends, and implement preventive measures. Such web-based tools can also serve as educational platforms for non-experts, offering dynamic content that conveys critical health information in an accessible way.

- **For Peer-reviewed Journals**: A shift from static articles to interactive research presentations is already happening. Epidemiology journals that provide accompanying dynamic tools for interpreting datasets will engage a wider audience and increase the accessibility and impact of cutting-edge research.

---

**4. Gen Z is Pragmatic and Solutions-focused—Let’s Meet Them There**

Gen Z isn’t just looking for theoretical knowledge; they want solutions. Dynamic tools fit their problem-solving mindset and offer hands-on, practical applications of epidemiological theories and models. By integrating web apps into our curriculum, we are not only engaging students in a more meaningful way but also preparing them to address real-world public health crises using modern technology.

---

**Conclusion:**
Moving towards dynamic, interactive web apps will not only revolutionize our approach to teaching epidemiology but also place our department at the cutting edge of public health education. Let’s lead the way in adapting to the future of education by meeting the needs and preferences of the Gen Z learners who are already shaping the field of epidemiology.

---

This approach capitalizes on Gen Z’s strengths and pragmatism, emphasizing how dynamic tools will improve learning outcomes for students and professionals alike.
